Question 1: Block matrix multiplication

```python
import numpy as np
import requests
from bs4 import BeautifulSoup
import csv
import time
import pandas as pd
```

```python
def block_multiply(A, B, size_block):
    A = np.array(A)
    B = np.array(B)

    if A.shape[1] != B.shape[0]:
        raise ValueError("Invalid matrix dimensions")

    ans = np.zeros((A.shape[0], B.shape[1]))

    for i in range(0, A.shape[0], size_block):
        for j in range(0, B.shape[1], size_block):
            for k in range(0, A.shape[1], size_block):
                ans[i:i+size_block, j:j+size_block] += np.dot(
                    A[i:i+size_block, k:k+size_block], B[k:k+size_block, j:j+size_block]
                )

    return ans

A = [
    [1, 2, 4, 1],
    [1, 0, -1, -2],
    [2, -1, 3, 1],
    [1, 2, 3, 4]
]

B = [
    [-2, 2, -3, 3],
    [0, 1, -1, 2],
    [-2, -1, 0, 1],
    [4, 0, 1, 4]
]

size_block = 2

result = block_multiply(A, B, size_block)

print("Result of Block Matrix Multiplication:")
for row in result:
    print(row)
```

Question 3:

Construct a dataset by scraping a
website online. Utilize BeautifulSoup or any other suitable scraping library to extract all the
details of Amazon products and compile a dataset.

```python
def extract_product_title(soup):
   title_element = soup.find("span", id="productTitle")
   if title_element:
       return title_element.text.strip()
   else:
       return ""

def extract_product_rating(soup):
    try:
        rating_tag = soup.find("i", attrs={'class': 'a-icon a-icon-star a-star-4-5'}).string.strip()

    except AttributeError:
        try:
            rating_tag = soup.find("span", attrs={'class': 'a-icon-alt'}).string.strip()
        except:
            rating_tag = ""

    return rating_tag
def extract_review_count(soup):
   review_count_element = soup.find("span", id="acrCustomerReviewText")
   if review_count_element:
       return review_count_element.string.strip()
   return ""

def extract_availability_status(soup):
   availability_element = soup.find("div", id="availability")
   if availability_element and availability_element.find("span").string:
       return availability_element.find("span").string.strip()
   else:
       return "Not Available"
```

```python
search_url = "https://www.amazon.com/s?k=playstation+4&ref=nb_sb_noss_2"
headers = ({'User-Agent':'', 'Accept-Language': 'en-US, en;q=0.5'})

response = requests.get(search_url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

product_data = {"title": [], "rating": [], "reviews": [], "availability": []}

product_links = soup.find_all("a", class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal")
product_links = [link.get("href") for link in product_links]

for product_link in product_links:

       new_webpage = requests.get("https://www.amazon.com" + product_link, headers=headers)

       product_soup = BeautifulSoup(new_webpage.content, "html.parser")

       # Extract product details using provided functions
       product_data["title"].append(extract_product_title(product_soup))
       product_data["rating"].append(extract_product_rating(product_soup))
       product_data["reviews"].append(extract_review_count(product_soup))
       product_data["availability"].append(extract_availability_status(product_soup))

   # Create and clean DataFrame
df = pd.DataFrame.from_dict(product_data)
df['title'].replace('', np.nan, inplace=True)
df = df.dropna(subset=['title'])

df.to_csv("amazon_data.csv", header=True, index=False)
df
```
