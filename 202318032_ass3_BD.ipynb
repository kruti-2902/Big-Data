```
file = rdd.collect()
```

```
import re
# print(data)
pattern = r'[\n\s!@#$%^&*\(\)\{\}\[\]~`/\?<>,\.]+'
words = re.split(pattern, str(file))
words
```

```
word_count = list(map(lambda x: (x, 1), words))
word_count
```

```
word_counts = {}
for word in words:
    word = word.lower()
    word_counts[word] = word_counts.get(word, 0) + 1

sorted_word_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

print(f'Total number of words: {sum(word_counts.values())}\n')
print('List of Words and their Counts (sorted by count in descending order):\n')
for word, count in sorted_word_counts:
    print(f'{word}: {count}')
```
