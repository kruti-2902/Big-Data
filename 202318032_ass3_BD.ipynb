```
file = rdd.collect()
```

```
import re
# print(data)
pattern = r'[\n\s!@#$%^&*\(\)\{\}\[\]~`/\?<>,\.]+'
words = re.split(pattern, str(file))
words
```

```
word_count = list(map(lambda x: (x, 1), words))
word_count
```

```
word_counts = {}
for word in words:
    word = word.lower()
    word_counts[word] = word_counts.get(word, 0) + 1

print(f'Total number of words:  {sum(word_counts.values())} \n\nList of Words:  \n\n{word_counts.items()} ')
```
