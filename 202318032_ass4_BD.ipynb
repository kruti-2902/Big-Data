TASK-2: 
    
In this task you will calculate the frequency of each word in text8 dataset mentioned above. Create a RDD using the text8 dataset.Use appropriate functions of the RDD to get the word frequencies. Filter the RDD using appropriate function to get the frequencies of words containing the letter ’a’.

```
text8 = sc.textFile("dbfs:/FileStore/shared_uploads/202318032@daiict.ac.in/text8")

# Split the text into individual words
words = text8.flatMap(lambda line: line.lower().split())

# Calculate the frequency of each word
frequency = words.countByValue()

# Filter the frequencies to select words containing the letter 'a'
words_with_a_freq = {word: freq for word, freq in frequency.items() if 'a' in word}

for word, freq in words_with_a_freq.items():
    print(f"Word: {word}, Frequency: {freq}")
```

DataFrame Task: 

Create a Spark dataframe using the iris json data mentioned above. Calculate Pearson Correlation between the columns petalLength and petalWidth using the appropriate dataframe API. Show the columns sepalLength, sepalWidth and species for the rows of data that has petalLength greater than or equal to 1.4 using the appropriate dataframe API.

```
from pyspark.sql.functions import corr
```

```
iris = spark.read.json("dbfs:/FileStore/shared_uploads/202318032@daiict.ac.in/iris.json")

correlation = iris.stat.corr("petalLength", "petalWidth")
print(f"Pearson Correlation between petalLength and petalWidth: {correlation}")
```

```
df = iris.filter("petalLength >= 1.4").select("sepalLength", "sepalWidth", "species")
df.show()
```
