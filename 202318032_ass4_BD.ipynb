TASK-2: 
    
In this task you will calculate the frequency of each word in text8 dataset mentioned above. Create a RDD using the text8 dataset.Use appropriate functions of the RDD to get the word frequencies. Filter the RDD using appropriate function to get the frequencies of words containing the letter ’a’.

```
text8_rdd = sc.textFile("dbfs:/FileStore/shared_uploads/202318032@daiict.ac.in/text8")

# Split each line into words and flatten the list of words
words = text8_rdd.flatMap(lambda line: line.split())

# Map each word to a tuple of (word, 1) for counting
word_count = words.map(lambda word: (word, 1))

# Reduce by key to get word frequencies
word_frequencies = word_count.reduceByKey(lambda x, y: x + y)

# Filter out words containing the letter 'a'
words_a = word_frequencies.filter(lambda pair: 'a' in pair[0])

# Collect results
word_frequencies_with_a = words_a.collect()

# Print word frequencies containing 'a'
for word, frequency in word_frequencies_with_a:
    print(f"Word: {word}, Frequency: {frequency}")
```

DataFrame Task: 

Create a Spark dataframe using the iris json data mentioned above. Calculate Pearson Correlation between the columns petalLength and petalWidth using the appropriate dataframe API. Show the columns sepalLength, sepalWidth and species for the rows of data that has petalLength greater than or equal to 1.4 using the appropriate dataframe API.

```
from pyspark.sql.functions import corr
```

```
# Read iris JSON data into a Spark DataFrame
iris = spark.read.json("dbfs:/FileStore/shared_uploads/202318032@daiict.ac.in/iris.json")

# Calculate Pearson Correlation between petalLength and petalWidth
pearson_corr = iris.select(corr("petalLength", "petalWidth")).collect()[0][0]
print("Pearson correlation between petalLength and petalWidth is :", pearson_corr)
```

```
data1 = iris.filter(iris["petalLength"] >= 1.4).select(iris["sepalLength"], iris["sepalWidth"], iris["species"])
data1.show()
```
